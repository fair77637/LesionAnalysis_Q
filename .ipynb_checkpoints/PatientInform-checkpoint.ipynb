```python
# for all the cases where the admission GCS score is high but patients did not survive, the 
# overestimation theory predicts the death.
# Check univariant linear regression on :
# https://www.scipy-lectures.org/packages/statistics/index.html
```

```python
PO_O['GOS'] = PO_O['GOS'].astype('int')
PO_O['GOSE'] = PO_O['GOSE'].astype('int')
```

```python
GCS = PO_O['GCS'][1:].astype('int')
GOS = PO_O['GOS'][1:]
```

```python
def survive(x):
    if x == 1:
        w = 'dead'
    else:
        w = 'alive'
    return w
```

```python
survive = GOS.apply(survive).values
survive_dummy = sm.categorical(survive)
survive_dummy[:,1:]
```

```python
x = np.linspace(0, 20, len(survive_dummy))
X = np.column_stack((x, survive_dummy[:,1:]))
```

```python
data = pd.DataFrame({'GCS':GCS,'GOS':GOS,'survive':survive})
```

```python
from statsmodels.formula.api import ols
model = ols('GOS~GCS',data).fit()
model.summary()
```
